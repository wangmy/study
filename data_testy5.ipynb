```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['font.family'] = ['Arial Unicode MS']

mask_data_clean = pd.read_csv('./data-5/mask_data_clean.csv')
month_order = mask_data_clean.groupby('月份')['订单量'].sum()
month_order.plot(kind='line', figsize=(6,6), title='各月订单总量')
```

```python
# 数据可视化： Tableau PowerBI Python
# Python可视化常用的第三方库：pandas、matplotlib、seaborn
# matplotlib 库相较于pandas以及seaborn更为底层，因此matplotlib库中的绘图函数、参数相对更多
# Python 数据分析‘三剑客’：matplotlib、numpy、pandas
```

```python
# 用matplotlib库绘图
# 图标元素：图标标题、坐标轴刻度、坐标轴标题、图例、数据标签
```

```python
# 生成画布、并设置大小
# 画布大小为 6 英寸x6 英寸，默认分辨率为 72 像素/英寸，所以像素为 432x432。
plt.figure(figsize=(6,6)) 

# 设置x/y
x = pd.Series(['第一季度','第二季度','第三季度','第四季度'])
y = pd.Series([59, 70, 68, 56])
plt.plot(x, y, color='dodgerblue')
```

```python
# 绘制柱状图
plt.bar(x, height=y, color='darkorange')
plt.title('各季度收入', fontsize=20)
# 设置x/y轴字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# 设置x/y标题
plt.xlabel('季度', fontsize=15)
plt.ylabel('营业收入（千万元）', fontsize=15)
# 设置图例
plt.legend(['变化','分布'])
# 设置数据标签
# plt.text(x,y,s, ha=None, va=None, fontsize=None)
    # x,y 数据标签的位置（数值或字符串）
    # s 数据标签的内容（数值或字符串）
    # ha 水平对齐方式（left、right、center）
    # va 垂直对齐方式（center、top、bottom）
    # fontsize 数据标签的字体大小（数值类型）

# plt.text('第三季度', 68, 689, ha='center', va='bottom', fontsize=12)

# 使用zip()来添加标签
for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
```

```python
# 绘制饼图
# 生成画布，并设置画布的大小
plt.figure(figsize=(6, 6))
# 设置扇形面积值
x = pd.Series([59, 70, 68, 56])
# 设置百分比小数的位数：保留百分比小数点后两位
autopct='%.2f%%'
# 设置百分比字体大小和颜色
textprops = {'fontsize': 12, 'color': 'black'}
# 设置饼图的“爆炸效果”：让扇形远离圆心
explode = [0.3, 0.1, 0, 0]
# 设置不同扇形的颜色
colors = ['cornflowerblue', 'salmon', 'yellowgreen', 'orange']
# 绘制饼图
plt.pie(x, autopct=autopct, textprops=textprops,
        explode=explode, colors=colors)
# 设置图表标题名及字体大小
plt.title('闪光科技2020年各季度研发费用占比饼图', fontsize=20)
# 设置图例
plt.legend(['第一季度', '第二季度', '第三季度', '第四季度']) 
```

```python
# 生成画布，并设置画布的大小为(6, 6)
plt.figure(figsize=(6,6))
# 设置 x/y坐标值
x = pd.Series(['第一季度', '第二季度', '第三季度', '第四季度'])
y = pd.Series([59, 70, 68, 56])
# 绘制柱状图
plt.bar(x,height=y,color='darkorange')
# 绘制折线图
plt.plot(x, y, color='dodgerblue')
# 设置图表标题名为'闪光科技2020年各季度研发费用分布图'，字体大小为 20
plt.title('闪光科技2020年各季度研发费用分布图', fontsize=20)
# 设置坐标轴的刻度字体大小为 12
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 设置 x、y 轴标题分别为'季度'、'研发费用（百万元）'，字体大小均为 15
plt.xlabel('季度', fontsize=15)
plt.ylabel('研发费用（百万元）', fontsize=15)

# 设置图例为'研发费用变化'、'研发费用分布'
plt.legend(['研发费用变化','研发费用分布'])

# 设置数据标签，标签内容是 y 坐标值
for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)

```
